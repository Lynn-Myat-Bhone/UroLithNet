```python
%%writefile data.yaml
path: /content/dataset

train: train/_annotations.coco.json
val: valid/_annotations.coco.json
test: test/_annotations.coco.json

names:
  0: kidney-stones
  1: Tas_Var
```

```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")  # better than nano for medical

model.train(
    data="data.yaml",
    epochs=180,
    imgsz=640,
    batch=16,
    optimizer="AdamW",
    lr0=0.001,
    weight_decay=0.0005,
    cos_lr=True,
    close_mosaic=15,
    patience=30,
    device=0
)
```

```python
metrics = model.val(data="data.yaml", split="test")
print(metrics)
```
